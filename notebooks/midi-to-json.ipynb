```python
import mido
import numpy as np
from mido import MidiFile
import matplotlib.pyplot as plt
import json
```

```python
mid= MidiFile('tu-jesty.mid')
```

```python
fs = 44100
H = 512
melody = []

for msg in mid.play():
    if msg.type == 'note_off':
        time = msg.time
        note = msg.note
        
        for i in range(int(time * fs / H)):
            melody.append(note)
```

```python
t = np.arange(len(melody)) * H / fs
plt.plot(t, melody)
```

```python
data = {
    'melody': melody
}

with open('tu-jesty.json', 'w') as fp:
    json.dump(data, fp, indent=4)
```
